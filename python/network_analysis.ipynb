# Network analysis for when the data is scraped

```python
import pandas as pd
import networkx as nx
```

```python
votes_data = pd.read_csv('lor_abridged.csv')   
    
G = nx.from_pandas_edgelist(votes_data, 
                            source='match_id',
                            target='player',
                            edge_attr=None,
                            create_using=nx.DiGraph())
                            
nx.draw_networkx(G)
```
