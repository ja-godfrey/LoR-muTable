```python
import pandas as pd
import json
import csv
import riotwatcher
from riotwatcher import LorWatcher, ApiError

# bring in existing databases
df_database = pd.read_csv('matches_database.csv')
df_puuids = pd.read_csv('puuids_database.csv')
df_match_ids = pd.read_csv('match_ids_database.csv')

# api stuff
api_key = 'wouldn\'t you like to know'
watcher = LorWatcher(api_key)
region = 'americas'
```

```python
# gets the match history. This is the json file with all the dictionaries.
def get_match_history(region, puuid): #define region with import stuff, puuid should be drawn from csv
    
    #get list of match_ids
    data = watcher.match.by_puuid(region=region, puuid=puuid) 
    
    #write new match_ids to a separate csv
    with open('match_ids_database.csv', "a") as fp:
        wr = csv.writer(fp, dialect='excel')
        for val in data:
            wr.writerow([val])
    
    #delete dupes
    df = pd.read_csv('match_ids_database.csv')
    df.drop_duplicates(inplace=True)
    df.to_csv('match_ids_database.csv', index=False)
    
    return data
```

```python
# turns the match history into a database that you can use
# takes the json as dict, not the match_history code
#YOU DON'T RUN THIS DIRECTLY. IT RUNS THROUGH THE FUNCTION BELOW.
def build_db_from_match_id(match_id):
    if match_id['info']['game_type'] == 'Ranked':
        df_0 = {}
        df_1 = {}

        # build dictionary for participant 0
        df_0['match_id'] = match_id['metadata']['match_id']
        df_0['participant'] = match_id['metadata']['participants'][0]
        df_0['game_start_time_utc'] = match_id['info']['game_start_time_utc']
        df_0['game_version'] = match_id['info']['game_version']
        df_0['deck_id'] = match_id['info']['players'][0]['deck_id']
        df_0['deck_code'] = match_id['info']['players'][0]['deck_code']
        df_0['faction0'] = match_id['info']['players'][0]['factions'][:1]
        if match_id['info']['players'][0]['factions'][1:]:
            df_0['faction1'] = match_id['info']['players'][0]['factions'][1:]
        else:
            df_0['faction1'] = ["N/A"]
        df_0['game_outcome'] = match_id['info']['players'][0]['game_outcome']
        df_0['order_of_play'] = match_id['info']['players'][0]['order_of_play']    

        # build dictionary for participant 1
        df_1['match_id'] = match_id['metadata']['match_id']
        df_1['participant'] = match_id['metadata']['participants'][1]
        df_1['game_start_time_utc'] = match_id['info']['game_start_time_utc']
        df_1['game_version'] = match_id['info']['game_version']
        df_1['deck_id'] = match_id['info']['players'][1]['deck_id']
        df_1['deck_code'] = match_id['info']['players'][1]['deck_code']
        df_1['faction0'] = match_id['info']['players'][1]['factions'][:1]
        if match_id['info']['players'][1]['factions'][1:]:
            df_1['faction1'] = match_id['info']['players'][1]['factions'][1:]
        else:
            df_1['faction1'] = ["N/A"]
        df_1['game_outcome'] = match_id['info']['players'][1]['game_outcome']
        df_1['order_of_play'] = match_id['info']['players'][1]['order_of_play']

        # convert dictionaries to dataframes
        df_0 = pd.DataFrame().from_dict(df_0)
        df_1 = pd.DataFrame().from_dict(df_1)

        # append dataframes
        dfXXX = df_0.append(df_1)
        return dfXXX
```

```python
# builds the matches database using the previous functions
def build_database(match_id_list, df_database): # takes match_ids generated from the start
    for i in match_id_list: 
        match_id = watcher.match.by_id(region=region, match_id=i) #calls api to get custom match json
        x = build_db_from_match_id(match_id) #uses custom function to turn json into dataframe
        df_database = df_database.append(x) # appends new stuff to existing dataframe
        df_database.drop_duplicates(inplace=True) # drops duplicate rows
        df_database.to_csv('matches_database.csv', index=False) #rewrites old database
    return df_database
```

```python
# builds puuid database
def get_puuids(df_database): # get all the puuids from the current database of matches
    
    #get list of puuids
    puuids = list(df_database['participant'])
    
    #write new puuids to a separate csv
    with open('puuids_database.csv', "a") as fp:
        wr = csv.writer(fp, dialect='excel')
        for val in puuids:
            wr.writerow([val])
    
    #delete dupes
    df = pd.read_csv('puuids_database.csv')
    df.drop_duplicates(inplace=True)
    df.to_csv('puuids_database.csv', index=False)
    
    return puuids
```
